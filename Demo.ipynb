```python
import warnings
warnings.filterwarnings("ignore")

import cv2
import matplotlib.pyplot as plt
from scipy.ndimage.interpolation import shift
from scipy.ndimage import zoom
from mi_reg import main_mi_reg
from poc_reg import main_poc_reg

%matplotlib inline
```

# Read original image (only one channel) and resize it

```python
im_orig = cv2.imread("image-test.jpg")
im_orig_0 = im_orig[:,:,0]
im_orig_0 = zoom(im_orig_0, zoom=0.2)
plt.imshow(im_orig_0, cmap='gray')
```

# Create a shifted version (5.3p in X, 6.7p in Y)

```python
im_shifted = shift(im_orig_0, [5.3, 6.7])
plt.imshow(im_shifted, cmap='gray')
```

```python
dx, dy, match_height = main_mi_reg(im_orig_0, im_shifted)
print(dx, dy, match_height)
```

```python
dx, dy, match_height = main_poc_reg(im_orig_0, im_shifted)
print(dx, dy, match_height)
```

# Both match heights are close to 1 indicating high level of confidence in detected  shifts
